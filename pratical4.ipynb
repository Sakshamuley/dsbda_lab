```python

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv(r"C:\Users\saksh\Downloads\bostonhousing-bbm.csv")
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
column_names = [
    'crim', 'zn', 'indus', 'chas', 'nox', 'rooms', 'age', 'distance',
       'radial', 'tax', 'pt', 'lstat', 'mvalue'
]
df.columns = column_names
```

```python

X = df.drop(['mvalue'], axis=1)
y = df['mvalue']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

lm = LinearRegression()
model = lm.fit(xtrain, ytrain)
```

```python

ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)

```

```python
print("Train MSE:", mean_squared_error(ytrain, ytrain_pred))
print("Test MSE:", mean_squared_error(ytest, ytest_pred))
print("Train R²:", r2_score(ytrain, ytrain_pred))
print("Test R²:", r2_score(ytest, ytest_pred))
```

```python

plt.scatter(ytrain, ytrain_pred, c='blue', marker='o', label='Training data')
plt.scatter(ytest, ytest_pred, c='lightgreen', marker='s', label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted values')
plt.title("True vs Predicted House Prices")
plt.legend(loc='upper left')
plt.show()
```
